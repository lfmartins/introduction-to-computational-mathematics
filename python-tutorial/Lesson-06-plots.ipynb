# Simple plots

Ploting using `matplotlib` is a three step process:

- Define the range of $x$ values using the `np.linspace()` function.
- Compute the corresponding $y$ values.
- Display the plot using the `plt.plot()` function.

As an example, let's consider the Fourier series for a square wave of period $2\pi$:

$$
\newcommand{\mtab}{\quad\quad\quad\quad}
\mtab f(x) = \frac{4}{\pi}\sum_{n=0}^{\infty}\frac{\sin((2n+1)x)}{2n+1}
$$

Suppose we want to see how good is the approximation given by the first three terms of the series. The following cell shows how to build the plot.

```python
plt.figure(figsize=(8,6))
xvalues = np.linspace(-2*np.pi, 2*np.pi, 400)
yvalues = 4/np.pi * (np.sin(xvalues) + np.sin(3*xvalues)/3 + np.sin(5*xvalues)/5)
plt.plot(xvalues, yvalues)
None
```

The assignment:

    xvalues = np.linspace(-2*np.pi, 2*np.pi, 400)
    
stores in `xvalues` an array of 200 equally spaced points in the interval $[0,2\pi$]. Here are the first five values:

```python
xvalues[:5]
```

We then need to compute the corresponding array `yvalues` of the y-values to be plotted. This is done with the expression:

    4/np.pi * (np.sin(xvalues) + np.sin(3*xvalues)/3 + np.sin(5*xvalues)/5)
    
Now we want to plot the square wave to compare with the Fourier approximation. The following cell shows how this can be done:

```python
xvalues = [-2*np.pi, -np.pi, -np.pi,  0, 0, np.pi, np.pi, 2*np.pi]
yvalues = [1, 1, -1, -1, 1, 1, -1, -1]
plt.plot(xvalues, yvalues)
None
```

For the square wave plot, it is easier to just enter the points that define the graph manually, instead of using a formula.

In the next cell, we plot both the Fourier approximation and the square wave for comparison.

```python
plt.figure(figsize=(8,6))
xvalues = np.linspace(-2*np.pi, 2*np.pi, 400)
yvalues = 4/np.pi * (np.sin(xvalues) + np.sin(3*xvalues)/3 + np.sin(5*xvalues)/5)
plt.plot(xvalues, yvalues)
xvalues = [-2*np.pi, -np.pi, -np.pi,  0, 0, np.pi, np.pi, 2*np.pi]
yvalues = [1, 1, -1, -1, 1, 1, -1, -1]
plt.plot(xvalues, yvalues)
None
```

`Matplotlib` automatically selects the colors of the plots. We will se later how to choose colors explicitly.

The syntax of the plot command is:

    plot(xvalues, yvalues, options)
    
The arrays `xvalues` and `yvalues` must have the same lengths. `options` are formatting specifications for the plot.

Let's now improve the graph with labels for the axes, a title, and centered axes. We also show how to control de color and line width of the plots.

```python
plt.figure(figsize=(8,6))
# Draw centered axes
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

# Draw graphs
xvalues = np.linspace(-2*np.pi, 2*np.pi, 400)
yvalues = 4/np.pi * (np.sin(xvalues) + np.sin(3*xvalues)/3 + np.sin(5*xvalues)/5)
plt.plot(xvalues, yvalues, color='green', lw=2)
xvalues = [-2*np.pi, -np.pi, -np.pi,  0, 0, np.pi, np.pi, 2*np.pi]
yvalues = [1, 1, -1, -1, 1, 1, -1, -1]
plt.plot(xvalues, yvalues, color='red', lw=2)

# Add title and labels
plt.title('Four-term Fourier approximation for a square wave')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
None
```
